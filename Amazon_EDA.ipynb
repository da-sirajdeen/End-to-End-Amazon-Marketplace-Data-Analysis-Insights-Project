Charts - Insights

```python
##Bar chart: Revenue by category

##Visualize total revenue contribution across different product categories to identify top-performing segments.


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('amazon_ecommerce_1M.csv') 

revenue_by_cat = df.groupby('category')['final_price'].sum().sort_values(ascending=False).reset_index()
revenue_by_cat.columns = ['category', 'total_revenue']


plt.figure(figsize=(10, 6))
sns.barplot(data=revenue_by_cat, x='category', y='total_revenue', palette='Blues_d',color='blue')

plt.title('Total Revenue by Category', fontsize=16)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Total Revenue', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
##Understand how product prices and discounts are distributed across the dataset.

fig, axes = plt.subplots(1, 2, figsize=(14, 6))


sns.histplot(df['final_price'], bins=40, color='steelblue', ax=axes[0], kde=True)
axes[0].set_title('Price Distribution', fontsize=14)
axes[0].set_xlabel('Final Price', fontsize=12)
axes[0].set_ylabel('Number of Orders', fontsize=12)


sns.histplot(df['discount'], bins=40, color='navy', ax=axes[1], kde=True)
axes[1].set_title('Discount Distribution', fontsize=14)
axes[1].set_xlabel('Discount (0 = 0%, 0.5 = 50%)', fontsize=12)
axes[1].set_ylabel('Number of Orders', fontsize=12)

plt.suptitle('Price & Discount Distribution', fontsize=16, y=1.02)
plt.tight_layout()
plt.show()
```

```python

##Understand the relationships between numerical variables in the dataset.

numeric_df = df.select_dtypes(include='number')


correlation = numeric_df.corr()


plt.figure(figsize=(12, 8))
sns.heatmap(correlation, 
            annot=True,
            fmt='.2f',  
            cmap='coolwarm', 
            linewidths=0.5, 
            square=True)       

plt.title('Correlation Heatmap', fontsize=16)
plt.tight_layout()
plt.show()
```

```python

##Convert the dataset into a structured database format for efficient storage and SQL-based querying.


import pandas as pd
import sqlite3

df = pd.read_csv('amazon_ecommerce_1M.csv')


conn = sqlite3.connect('ecommerce.db')


df.to_sql('orders', conn, if_exists='replace', index=False)

print("Database ready!")
```
